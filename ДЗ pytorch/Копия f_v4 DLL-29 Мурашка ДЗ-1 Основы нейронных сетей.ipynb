```python
batch_size = 10

dataset_train = TensorDataset(torch.from_numpy(X_train.values).type(torch.float), torch.from_numpy(y_train.values).type(torch.float))
dataloader_train   = DataLoader(dataset_train, batch_size, shuffle=True)

dataset_test  = TensorDataset(torch.from_numpy(X_test.values).type(torch.float), torch.from_numpy(y_test.values).type(torch.float))
dataloader_test  = DataLoader(dataset_test, batch_size)
```

```python
loss = torch.nn.MSELoss(reduction='mean')
```

```python
trainer = torch.optim.SGD(model.parameters(), lr=0.000001)
```

```python
num_epochs = 300
for epoch in range(1, num_epochs + 1):
    train_iters, train_loss = 0., 0.
    model.train() # переводим в режим обучения
    for Xl, yl in dataloader_train:
        trainer.zero_grad()
        l = loss(model(Xl).reshape(-1), yl)
        l.backward()
        trainer.step()
        train_loss += l.item()
        train_iters += 1

    test_iters, test_loss = 0., 0.
    model.eval() # переводим в режим исполнения
    for Xt, yt in dataloader_test:
        l = loss(model(Xt).reshape(-1), yt)
        test_loss += l.item()
        test_iters += 1

    if epoch % 5 == 0:
 #      print('epoch %d, loss: %f' % (epoch, l.item()))
        print('epoch %d, loss: %f' % (epoch, train_loss / train_iters),' test_loss: %f' % (test_loss / test_iters),'|\tw', model[0].weight.data, '|\tb', model[0].bias.data)
```
